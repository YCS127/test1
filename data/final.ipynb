## Préparation data

```python
# sélection des variables qui participent aux analyses
#data = data_import
#data = data_import.iloc[:,1:14]

data = data_import.iloc[:,[2, 3, 4, 5, 6, 7]]
print ("n = {} individus (pays) qualifiés par p = {} dimensions (variables numériques)"
       .format(data.shape[0], data.shape[1]))
print("")
print("Intitulé des {} variables numériques".format(data.shape[1]))
print(list(data.columns))
```

```python
from sklearn import preprocessing
# les data
X = data.values
names = data.index
# Centrage et Réduction
X_scaled = preprocessing.StandardScaler().fit_transform(X)
```

## Recherche des données aberrantes (model: IsolationForest)

```python
# détection anomalie
from sklearn.ensemble import IsolationForest
#choix du model
model = IsolationForest(contamination=0.02)
#entrainement du model
model.fit(X_scaled)

# on selectionne les outliers potentiels
outliers_select = (model.predict(X_scaled) == -1)
        
df = pd.DataFrame({
    "F1" : X_scaled[:,0], 
    "F2" : X_scaled[:,1],
    "Zone" : data.index,
    "Rt" : outliers_select
})

# utilisation de subplots nécessaire car annotation du graphique
fig, ax = plt.subplots(figsize=(5, 5))
df.plot.scatter("F1", "F2", ax = ax) # l'option ax permet de placer les points et le texte sur le même graphique

# boucle sur chaque pays
for k in df.iterrows():
    # annotation uniquement si valeur absolue sur une de 2 dimensions importantes (valeurs choisies empiriquement)
    if (k[1]['Rt'] == True):
        ax.annotate(k[1]["Zone"], (k[1]['F1'], k[1]['F2']), fontsize = 11, c='red')
plt.title('IsolationForest model', fontsize=15)
plt.xlabel('')
plt.ylabel('')
text_str = 'Pas de outliers, seulement des valeurs extrêmes'
plt.text(5.5, 0, text_str, horizontalalignment = 'left', verticalalignment = 'center', style = 'italic', alpha = 0.7, 
        fontsize = 12, backgroundcolor = 'lightgray') 
plt.show()  


```

## Nombre de clusters adapté

- Nous utiliserons d'abord la CAH afin d'envisager un nombre de clusters (il n'est pas nécessaire de fixer à priori un nombre de clusters pour cet algorithme)

- Nous vérifierons ce résultat avec les algorithmes k-means et ACP

### Classification hiérarchique ascendante (CAH)

- Dendrogramme

```python
# importation librairie
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

#générer la matrice des liens
Z = linkage(X_scaled,method='ward',metric='euclidean')
#affichage du dendrogramme
fig, ax = plt.subplots(figsize=(9,9))
plt.title("CAH (dendrogramme)", fontsize=15)
dendrogram(Z,labels=names,orientation='left',color_threshold=0)
plt.show()
```

- Diagramme d'inertie

```python
# diagramme d'inertie (scree plot) pour définir le nombre optimal de classes
last = Z[-10:, 2]
last_rev = last[::-1]
idxs = np.arange(2, len(last) + 2)
fig, ax = plt.subplots(figsize=(6, 6))
plt.step(idxs, last_rev, c="black")
plt.xlabel("Nombre de classes")
plt.ylabel("Inertie")
plt.title('Diagramme d\'inertie (scree plot)', fontsize=15)
text_str = 'L\'option 4 clusters paraît intéressante'
plt.text(12, 15, text_str, horizontalalignment = 'left', verticalalignment = 'center', style = 'italic', alpha = 0.7, 
        fontsize = 12, backgroundcolor = 'lightgray') 

nombre = 4 # Ici, on teste le scénario où il y aurait 4 catégories

plt.scatter(idxs[np.where(idxs==nombre)], last_rev[np.where(idxs==nombre)], c="red")
plt.axvline(idxs[np.where(idxs==nombre)], c="red")
plt.show()
```

- Dendrogramme colorisé (4 clusters)

```python
# récupération de labels y de la CAH (pour des comparaisons futures)
from sklearn.cluster import AgglomerativeClustering
y_pred = AgglomerativeClustering(n_clusters = 4).fit_predict(X_scaled)
```

```python
#matérialisation des 4 classes (hauteur t = 17)
fig, ax = plt.subplots(figsize=(10,10))
plt.title('CAH avec matérialisation des 4 classes', fontsize=15)

dendrogram(Z,labels=names,orientation='left',color_threshold=17)
plt.show()
```

Vérifions ces résultats avec k-means et ACP

### K-means et ACP

- Coefficient de silhouette (coefficient d'homogénéité des clusters)

```python
from sklearn import cluster, metrics   # pour clusteriser et calculer les coef. de silhouette
from sklearn.cluster import KMeans

# on fixe le nombre de clusters
num_clusters = 4
                              
# k-means pour plusieurs taille de clusters (on va calculer le coef. de silhouette pour la taille de cluster la + appropriée)
silhouettes = []
for num_clusters in range(2, 10):
    cls = cluster.KMeans(n_clusters = num_clusters, n_init=1, init='k-means++')
    cls.fit(X_scaled) # on entraile le modèle
    silh = metrics.silhouette_score(X_scaled, cls.labels_)
    silhouettes.append(silh)
    
# on affiche les coef. de silhouette pour toutes les valeurs de clusters
fig, ax = plt.subplots(figsize=(6, 6))
plt.plot(range(2, 10), silhouettes, marker='o')
plt.title('Coefficient de silhouette \npour un nombre de clusters différent')
text_str = 'On privilégie le nombre de clusters pour lequel \nle coefficient de silhouette est le plus élevé (3 ou 4)'
plt.text(10, 0.5, text_str, horizontalalignment = 'left', verticalalignment = 'center', style = 'italic', alpha = 0.7, 
        fontsize = 14, backgroundcolor = 'lightgray') 
plt.show()


kmeans_color = cls.predict(X_scaled)
```

On regarde maintenant à quoi ressemblent les clusters.   
On projette les individus grâce à une ACP en comparant les options 3 ou 4 clusters

```python
from sklearn import decomposition

# on visualise pour 3 ou 4 clusters
fig = plt.figure(figsize=(12, 5))


# model
pca = decomposition.PCA(n_components = 4)
pca.fit(X_scaled)
X_projeted = pca.transform(X_scaled)

# % de variances expliquée en cumul
var_cumul = pca.explained_variance_ratio_

# option 3 clusters
cls3 = cluster.KMeans(n_clusters=3)
cls3.fit(X_scaled)

#option 4 clusters
cls4 = cluster.KMeans(n_clusters=4)
cls4.fit(X_scaled)

ax = fig.add_subplot(121)
ax.scatter(X_projeted[:,0], X_projeted[:,1], c=cls3.labels_)
#plt.scatter(cls3.cluster_centers_[:, 0], cls3.cluster_centers_[:, 1], marker='X', c='cyan')
ax.set_title("3 clusters", fontsize=15)
# nom des axes, avec le pourcentage d'inertie expliqué
plt.xlabel('F1 ({}%)'.format(round(100 * var_cumul[0], 0)))
plt.ylabel('F2 ({}%)'.format(round(100 * var_cumul[1], 0)))


ax = fig.add_subplot(122)
ax.scatter(X_projeted[:,0], X_projeted[:,1], c=cls4.labels_)
#plt.scatter(cls4.cluster_centers_[:, 0], cls4.cluster_centers_[:, 1], marker='X', c='cyan')
ax.set_title("4 clusters", fontsize=15)
plt.xlabel('F1 ({}%)'.format(round(100 * var_cumul[0], 0)))
plt.ylabel('F2 ({}%)'.format(round(100 * var_cumul[1], 0)))
plt.text(12, 0.5, 'Je choisis l\'option 4 clusters \n(Vénézuéla, cluster à lui tout seul!)', 
         horizontalalignment = 'left', verticalalignment = 'center', style = 'italic', alpha = 0.7, 
        fontsize = 12, backgroundcolor = 'lightgray')
plt.show()
```

- Comparons les résultats obtenus (option 4 clusters)  
entre la CAH, le k-means et la l'ACP

```python
# comparaison clustering CAH, k-means et ACP 
fig = plt.figure(figsize=(12, 5))

cls4 = cluster.KMeans(n_clusters=4)
cls4.fit(X_scaled)


ax = fig.add_subplot(121)
ax.scatter(X_projeted[:,0], X_projeted[:,1], c=cls4.labels_)
plt.xlabel('F1 ({}%)'.format(round(100 * var_cumul[0], 0)))
plt.ylabel('F2 ({}%)'.format(round(100 * var_cumul[1], 0)))
ax.set_title("4 clusters (K-means)", fontsize=15)

ax = fig.add_subplot(122)
ax.scatter(X_projeted[:,0], X_projeted[:,1], c=y_pred)
# nom des axes, avec le pourcentage d'inertie expliqué
ax.set_title("4 clusters (CAH)", fontsize=15)

# on quantifie la ressemblane entre les 2 clustering (k-means et CAH)
tx = str(round(metrics.adjusted_rand_score(cls4.labels_, y_pred)*100,0))
text_str = 'Taux de ressemblance {} %'.format(tx)
# nom des axes, avec le pourcentage d'inertie expliqué
plt.xlabel('F1 ({}%)'.format(round(100 * var_cumul[0], 0)))
plt.ylabel('F2 ({}%)'.format(round(100 * var_cumul[1], 0)))
plt.text(12, 0.5, text_str, horizontalalignment = 'left', verticalalignment = 'center', style = 'italic', alpha = 0.7, 
        fontsize = 12, backgroundcolor = 'lightgray') 

plt.show()
```

## Etude des différents clusters (à la recherce de notre cible)

### Préparation data

```python
# identification des 4 groupes obtenus
groupes_cah = fcluster(Z,t=17,criterion='distance')

# index triés des groupes
idg = np.argsort(groupes_cah)

# observations et leurs groupes
group = pd.DataFrame(data.index[idg],groupes_cah[idg])
group['Cluster'] = group.index
group.set_index('Zone', inplace=True)
data_group = pd.merge(data, group, on='Zone', how='left')

# liste pays par cluster
list1 = list(data_group[data_group['Cluster'] == 1].index)
list2 = list(data_group[data_group['Cluster'] == 2].index)
list3 = list(data_group[data_group['Cluster'] == 3].index)
list4 = list(data_group[data_group['Cluster'] == 4].index)

# on tranforme les valeurs en ranking pour comparer
data_group['export_rk'] = data_group['Exportations volailles (milliers de tonnes)'].rank(ascending=False)
data_group['prod_rk'] = data_group['Production volailles (milliers de tonnes)'].rank(ascending=False)
data_group['pib_rk'] = data_group['PIB/h ($)'].rank(ascending=False)
data_group['stab_rk'] = data_group['Indice_stabilité_politique'].rank(ascending=False)
data_group['infla_rk'] = data_group['Inflation (%)'].rank()
data_group['import_rk'] = data_group['Importations volailles (milliers de tonnes)'].rank(ascending=False)

# on ne conserve que les colonnes ranking
data_rk = data_group.iloc[:, 6:]
data_rk = data_rk.reindex(columns=['Cluster', 'pib_rk', 'stab_rk', 'infla_rk', 'prod_rk','export_rk','import_rk'])

# on supprime la colonne Cluster
data_gr1 = data_rk[data_rk['Cluster'] == 1].iloc[:, 1:]
data_gr2 = data_rk[data_rk['Cluster'] == 2].iloc[:, 1:]
data_gr3 = data_rk[data_rk['Cluster'] == 3].iloc[:, 1:]
data_gr4 = data_rk[data_rk['Cluster'] == 4].iloc[:, 1:]
```
